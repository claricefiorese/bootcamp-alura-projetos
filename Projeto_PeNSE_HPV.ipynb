# Entenda o que  o HPV 

## O que  HPV?

Sigla em ingls para Papilomavrus Humano (Human Papiloma Virus - HPV). Os HPV so vrus capazes de infectar a pele ou as mucosas. Existem mais de 150 tipos diferentes de HPV, dos quais 40 podem infectar o
trato genital. Destes, 12 so de alto risco e podem provocar cncer (so
oncognicos) e outros podem causar verrugas genitais. 

## Como o HPV  transmitido?

O vrus HPV  altamente contagioso, sendo possvel contaminar-se
com uma nica exposio, e a sua transmisso acontece por contato direto com a pele ou mucosa infectada. A principal forma  pela via sexual,
que inclui contato oral-genital, genital-genital ou mesmo manual-genital. Portanto, o contgio com o HPV pode ocorrer mesmo na ausncia de
penetrao vaginal ou anal. Tambm pode haver transmisso durante o
parto. Embora seja raro, o vrus pode propagar-se tambm por meio de
contato com a mo.
<br>Como muitas pessoas portadoras do HPV no apresentam nenhum sinal ou sintoma, elas no sabem que so portadoras do vrus, mas podem
transmiti-lo.

## Como se prevenir da transmisso do HPV?

As medidas de preveno mais importantes so:
- Uso do preservativo (camisinha) nas relaes sexuais.  importante ressaltar que o seu uso, apesar de prevenir a maioria das DSTs, no impede totalmente a infeco pelo HPV, pois, frequentemente as leses esto presentes em reas no protegidas pela camisinha. Na presena de infeco na vulva, na regio pubiana, perineal e perianal ou na bolsa escrotal, o HPV poder ser transmitido apesar do uso do preservativo. A camisinha feminina, que cobre tambm a vulva, evita mais eficazmente o contgio se utilizada desde o incio da relao sexual.
- Evitar ter muitos parceiros ou parceiras sexuais.
- Realizar a higiene pessoal.
- Vacinar-se contra o HPV.

## Definio das funes

```
#Criando a funo para criar um dicionrio com as respostas
def cria_dicionario_respostas(cod_pergunta: str, dicionario = dicionario):
  '''
  Esta funo cria um dicionrio com as respostas a partir do cdigo da pergunta

  Parmetros: 
    cod_pergunta: string com o cdigo da pergunta
    dicionario: DataFrame com os dados do arquivo "Dicionario_PENSE_Microdados_AMOSTRA2.xls"
  
  Retorna:
    pergunta: string com o texto da pergunta
    dicionario_respostas: dicionrio contendo as respostas possveis
  '''

  #Fazendo a query
  busca = dicionario.query(f"VARIVEL == '{cod_pergunta}'")

  #Selecionando o ndice da pergunta 
  indice_pergunta = (busca.index).tolist()[0]

  #Guardando o texto da pergunta
  #df.iloc[<linhas>, <colunas>]
  pergunta = dicionario.iloc[indice_pergunta, 1]

  #Criando um dicionrio para as respostas possveis
  dicionario_respostas = {}

  #Enquanto no encontro o cdigo da prxima pergunta
  indice = indice_pergunta + 1

  while True:

    valor = dicionario.loc[indice]["VARIVEL"]

    #Para corrigir a ocorrncia de "0 a 2999"
    if type(valor) is int or valor.replace(" a ", "").isnumeric():

      #Adicionando as variveis e as alternativas de resposta ao dicionrio
      nova_linha = {int(valor) : dicionario.iloc[indice]["QUESTIONRIO DO ALUNO"]}

      #Atualizando o dicionrio com a nova linha
      dicionario_respostas.update(nova_linha)

      indice += 1
    
    else:
      break
  
  return pergunta, dicionario_respostas
```

```
#Criando a funo para criar a tabela de frequncia
def cria_tabela_frequencia(cod_pergunta: str, dados: pd.DataFrame):
  '''
  Esta funo cria uma tabela de frequncia das respostas a partir do cdigo da pergunta
  
  Parmetros:
    cod_pergunta: string com o cdigo da pergunta
    dados: DataFrame com os dados do arquivo "PENSE_AMOSTRA2_ALUNO.CSV" a serem utilizados
  
  Retorna:
    tabela_frequencia: DataFrame, tabela de frequncia das respostas
 '''

  #Criando o dicionrio com as respostas e pegando index da pergunta no dicionario
  pergunta, dicionario_respostas = cria_dicionario_respostas(cod_pergunta)

  #Calculando a frequncia
  frequencia = round(dados[cod_pergunta].value_counts(), 2)

  #Calculando a porcentagem
  percentual = round(dados[cod_pergunta].value_counts(normalize = True) * 100, 2)

  #Criando o DataFrame
  tabela_frequencia = pd.DataFrame({"Frequncia" : frequencia, "Percentual" : percentual})

  #Renomeando o ndice
  tabela_frequencia.rename(index=dicionario_respostas, inplace=True)

  #Renomenado as variveis
  tabela_frequencia.rename_axis(pergunta, axis=1, inplace=True)

  return tabela_frequencia
```

```
#Criando a funo para criar a tabela da frequncia cruzada
def cria_frequencia_cruzada(dados: pd.DataFrame, cod_indice: str, cod_coluna: str, normaliza = ""):
  '''
  Esta funo cria uma tabela de frequncia cruzada a partir do cdigo da perguntas
  
  Parmetros:
    cod_indice: string com o cdigo da pergunta das linhas
    cod_coluna: string com o cdigo da pergunta das colunas
    normaliza: "columns", "index" ou True
    dados: DataFrame com os dados do arquivo "PENSE_AMOSTRA2_ALUNO.CSV" a serem utilizados
  
  Retorna:
    frequencia_cruzada: DataFrame, tabela da frequncia cruzada das variveis
 '''

  #Criando os dicionrios
  pergunta_indices, respostas_indices = cria_dicionario_respostas(cod_indice)
  pergunta_colunas, respostas_colunas = cria_dicionario_respostas(cod_coluna)

  #Calculando a frequncia cruzada
  frequencia_cruzada = pd.crosstab(dados[cod_indice], dados[cod_coluna], normalize=normaliza) * 100

  #Renomeando os ndices e as colunas
  frequencia_cruzada.rename(index = respostas_indices, inplace = True)
  frequencia_cruzada.rename(columns = respostas_colunas, inplace = True)
  frequencia_cruzada.rename_axis(pergunta_indices, axis=0, inplace = True)
  frequencia_cruzada.rename_axis(pergunta_colunas, axis=1, inplace = True)

  return frequencia_cruzada.round(2)
```

```
def draw_brace(ax, xspan, yy, text):
    """Draws an annotated brace on the axes."""
    xmin, xmax = xspan
    xspan = xmax - xmin
    ax_xmin, ax_xmax = ax.get_xlim()
    xax_span = ax_xmax - ax_xmin

    ymin, ymax = ax.get_ylim()
    yspan = ymax - ymin
    resolution = int(xspan/xax_span*100)*2+1 # guaranteed uneven
    beta = 300./xax_span # the higher this is, the smaller the radius

    x = np.linspace(xmin, xmax, resolution)
    x_half = x[:int(resolution/2)+1]
    y_half_brace = (1/(1.+np.exp(-beta*(x_half-x_half[0])))
                    + 1/(1.+np.exp(-beta*(x_half-x_half[-1]))))
    y = np.concatenate((y_half_brace, y_half_brace[-2::-1]))
    y = yy + (.05*y - .01)*yspan # adjust vertical position

    ax.autoscale(False)
    ax.plot(x, y, color='black', lw=1)

    ax.text((xmax+xmin)/2., yy+.07*yspan, text, ha='center', va='bottom', fontsize=12)

#Fonte: https://stackoverflow.com/questions/18386210/annotating-ranges-of-data-in-matplotlib
```

Estudantes com <font color = mediumvioletred>idade menor que 13 anos</font> no responderam as perguntas relacionadas  sade sexual e reprodutiva.


>Segundo o estudo [Vacina para HPV e comportamentos sexuais de risco](https://brsa.org.br/wp-content/uploads/wpcf7-submissions/1111/IDENTIFICADO-Vacina-para-HPV-e-in%C3%ADcio-da-vida-sexual-uma-avalia%C3%A7%C3%A3o-utilizando-regress%C3%A3o-em-descontinuidade-comentado-2.pdf), o risco de infeco pelo vrus HPV aumenta conforme o comportamento de risco e o nmero de parceiros sexuais.

> Alm disso, existem controvrsias em torno da vacinao contra o HPV. 
- A autorizao da vacinao pelos pais pode ser interpretada como uma autorizao do incio da vida sexual dos estudantes.
- Sendo a vacina a medida mais eficaz para se prevenir contra a infeco pelo HPV, as estudantes que j iniciaram a vida sexual podem deixar de proteger-se com preservativo por causa da vacina. 
- A vacina pode ser responsvel pela iniciao sexual precoce e ausncia de preservativo na primeira relao sexual.
